```python
import pandas as pd
```

```python
def data_clean(link,skip_rows):
    
    df=pd.read_csv(link,skiprows=skip_rows)
    #df=pd.read_json(link)
    #df=pd.read_excel(link)
    #df=pd.read_html(link)
    df.dropna(how='all',inplace=True)
    df.drop_duplicates(inplace=True)
    return df
```

```python
df=data_clean(r"C:\Users\Osamah\code python\Data Analysis\My_Orders.csv",0)
```

```python
#for see all the data and some fun
df.describe()
```

```python
country_city=df.groupby(['City'])[['Quantity','Total price']].sum()
country_city
```

```python
city_Q=df.groupby(['City'])['Quantity'].sum()
city_Q
```

```python
city_Q.plot(kind='barh',title='The sumery for the citys',ylabel='City',xlabel='Quantity',color='r')
```
