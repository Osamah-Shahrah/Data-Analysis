```python
import pandas as pd
```

```python
def data_clean(link,skip_rows):
    
    df=pd.read_csv(link,skiprows=skip_rows)
    #df=pd.read_json(link)
    #df=pd.read_excel(link)
    #df=pd.read_html(link)
    df.dropna(how='all',inplace=True)
    df.drop_duplicates(inplace=True)
    return df
```

```python
df=data_clean(r'C:\Users\Osamah\code python\Data Analysis\Orders.csv',0)
df
```

```python
#upper the frist chart
df['City']=df['City'].str.title()
```

```python
df['Branch']=df['Branch'].str.upper()
df['Email']=df['Email'].str.lower()
```

```python
df['Phone Number']=df['Phone Number'].str.replace('Tel: ','')
```

```python
df['Unit Price']=df['Unit Price'].round(0)
df['Discount']=df['Discount'].round(0)
```

```python
#add column to the df just write name the column and add value that well add in the last the table
df['Total price']=df['Quantity']*df['Unit Price']
```

```python
df['Discount']=df['Discount'].fillna(0)
```

```python
df['Cost']=df['Total price']-df['Discount']
```

```python
#other way to insert column for the df
df.insert(9,'Customer full Name',df['Customer First Name']+' '+df['Customer Last Name'])
```

```python
df.insert(5,'Branch Name',df['Branch'].str[:2])
```

```python
#the condition for the chang the data like if loc[condition,column return for you and can create new column]=the value yoe need to add or change it
df.loc[df['Status']==True,'Item']
```

```python
df.loc[df['Status']==True,'Order Statue']='Done'
df.loc[df['Status']==False,'Order Statue']='Pending'
df
```

```python
# multi conditions use loc[(1_condition) & | (2_condition)...,column you need change or creat  ] =the value if you need it
df.loc[(df['Status']==False) & (df['Order Date']<'2023-9-1'),'Order Statue']='Canceled'
df[df['Order Statue']=='Canceled' ]
```

```python
#othe way to use condition for change data use apply this fun use fun other to do any thang
#df['column to apply condition for him'].apply(func-for the work) and save to new colum or change data column
def level(row):
    if row > 1000:
        return 'High'
    else:
        return 'Low'
df['Cost Level']=df['Cost'].apply(level)
df
```

```python
import numpy as np
```

```python
#this cod to use condition with numpy for the opration on the numbers np.where(condition,if true,if false)
df['Discount Description']=np.where(df['Discount']==0,'No Discount',df['Discount']/df['Cost'])
df
```
